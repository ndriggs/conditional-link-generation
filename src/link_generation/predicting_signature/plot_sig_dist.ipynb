```python
from sage.all import BraidGroup, Integer, Link
import matplotlib.pyplot as plt
import numpy as np
```

```python
B = BraidGroup(7)
sigs = []
components = []
for _ in range(1250) :
    for length in range(33,37) :
        braid_word = np.random.choice([i for i in range(-6,7) if i != 0], size=length)
        braid_word = [Integer(i) for i in braid_word]
        braid = Link(B(braid_word))
        sig = braid.signature()
        sigs.append(sig)
sigs = np.array(sigs)
```

```python
plt.hist(components, bins=max(components)-min(components)+1, density=True)
plt.xlabel('Signature')
plt.ylabel('Density')
plt.title('Number of components 2,500 random braids Length 35 with Braid Index 7')
plt.show()
```

```python
plt.hist(sigs, bins=max(sigs)-min(sigs)+1, density=True)
plt.xlabel('Signature')
plt.ylabel('Density')
plt.title(f'{len(sigs)} Random Braids of Length 33 to 36 with Braid Index 7')
plt.show()
```
